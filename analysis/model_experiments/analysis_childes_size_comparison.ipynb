# Comparing Dropout for large models

```python
import wandb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotnine as p9

p9.theme_set(p9.theme_light())
sns.set_theme(style="whitegrid")
# Set Default fig size
plt.rcParams['figure.figsize'] = [10, 6]
```

```python
api = wandb.Api()
runs = api.runs("zeb/childes-size-english")

data = []
for run in runs:
    config = run.config
    metrics = run.summary_metrics
    if 'num_train_examples' not in config:
        config['num_train_examples'] = config['dataset']['num_examples']
    data.append({**config, **metrics})

df = pd.DataFrame(data)

params_map = {435584 : '400k', 633856 : '600k', 832128 : '800k',  1228672 : '1M', 4816640 : '5M', 19070464: '19M'}
model_sizes_sorted = ['400k', '600k', '800k', '1M', '5M', '19M']
data_size_to_token_count = {2357 : 300000, 5494 : 700000, 14122 : 1800000, 23540 : 3000000, 54919 : 7000000, 141255 : 18000000, 165070 : 21096863}
data_size_to_word_count = {2357: 80000, 5494: 180000, 14122: 500000, 23540: 800000, 54919: 1800000, 141255: 5000000, 165070: 5600000}

df['model_size'] = df['num_params'].map(params_map)
df['num_train_tokens'] = df['num_train_examples'].map(data_size_to_word_count)
df = df[df['num_train_tokens'] != 5600000]

df['model_size'] = df['model_size'].astype('category')
df['num_train_tokens'] = df['num_train_tokens'].astype('category')
```

```python
# Plot bar plot of syntactic accuracy for different model sizes, grouped by dropout rate

for model_size in model_sizes_sorted:
    g = sns.catplot(
        data=df[df['model_size'] == model_size],
        x="num_train_tokens",
        y="eval/best_babyslm_test_lexical",
        hue="attn_pdrop",
        kind="bar",
        height=6,
        aspect=1.5,
    )

    g.set_axis_labels("Data Size", "Lexical Accuracy")
    g.set(ylim=(0.4, 0.9))
    plt.title(f"Lexical Accuracy by Data Size and Dropout for {model_size} Parameters")
    plt.show()

```

```python
# Plot bar plot of syntactic accuracy for different model sizes, grouped by dropout rate

for model_size in model_sizes_sorted:
    g = sns.catplot(
        data=df[df['model_size'] == model_size],
        x="num_train_tokens",
        y="eval/best_babyslm_syntactic",
        hue="attn_pdrop",
        kind="bar",
        height=6,
        aspect=1.5,
    )

    g.set_axis_labels("Data Size", "Syntactic Accuracy")
    g.set(ylim=(0.4, 0.9))
    plt.title(f"Syntactic Accuracy by Data Size and Dropout for {model_size} Parameters")
    plt.show()

```

```python
def plot_max_model_by_model_size(eval_criteria, criteria_name, line=False):

    # Plot lexical accuracy by data size, taking the max dropout rate for each model size
    plt_data = df.groupby(['model_size', 'num_train_tokens'])[eval_criteria].max().reset_index()
    plt_data = plt_data.to_dict()
    plt_data = pd.DataFrame(plt_data)
    plt_data.sort_values('model_size', key=lambda x: x.map({v: i for i, v in enumerate(model_sizes_sorted)}), inplace=True)
    plt_data = plt_data.reset_index(drop=True)
    plt_data['num_train_tokens'] = plt_data['num_train_tokens'].astype('category')
    plt_data['model_size'] = pd.Categorical(plt_data['model_size'], categories=model_sizes_sorted, ordered=True)
    plt_data[eval_criteria] = plt_data[eval_criteria].astype(float) * 100

    # Turn data sizes into human-readable labels, using 'k' for thousands and 'M' for millions
    x_labels = plt_data['num_train_tokens'].cat.categories
    x_labels = [f"{int(x/1000)}k" if x < 1_000_000 else f"{x/1_000_000:.1f}M" if x % 1_000_000 else f"{int(x/1_000_000)}M" for x in x_labels]

    if line:
        # Line plot with log scale
        g = sns.lineplot(data=plt_data, x='num_train_tokens', y=eval_criteria, hue='model_size')
        g.set(xscale="log")
        plt.title(f"{criteria_name} Accuracy by Data Size for Different Model Sizes")
        plt.show()

    else:
        # # Bar plot with linear scale
        # g = sns.catplot(data=plt_data, x="num_train_tokens", y=eval_criteria, hue="model_size", kind="bar", height=6, aspect=1.5)
        # g.set_axis_labels("Data Size", criteria_name)
        # g.set(ylim=(0.4, 0.9))
        # plt.title(f"{criteria_name} by Data Size for Different Model Sizes")
        # plt.show()
        #Use plotnine

        fnt_size = 25

        # Set to a gentle color palette
        p9.theme_set(p9.theme_light())

        p = (p9.ggplot(plt_data, p9.aes(x='num_train_tokens', y=eval_criteria, fill='model_size'))
            + p9.coord_cartesian(ylim=(40, 82), expand=False)
            + p9.geom_bar(stat='identity', position=p9.position_dodge(width=0.9), width=0.8)
            + p9.labs(x='Data Size (words)', y='Accuracy (%)')

            + p9.theme(figure_size=(12, 6),
                       text=p9.element_text(size=fnt_size),
                       legend_position='top',
                       legend_text=p9.element_text(size=fnt_size), 
                       axis_text=p9.element_text(size=fnt_size),
                       axis_title=p9.element_text(size=fnt_size),
                       axis_line=p9.element_line(size=1),
                        axis_ticks=p9.element_line(size=1),
                        # Remove top line
                        panel_border=p9.element_blank(),)

            + p9.guides(fill=p9.guide_legend(title='Model Size'))
            + p9.scale_fill_manual(values=sns.color_palette('rocket').as_hex())
            + p9.scale_x_discrete(labels=x_labels)
        )        

        p.show()
```

```python
plot_max_model_by_model_size('eval/best_babyslm_test_lexical', 'Lexical Accuracy')
plot_max_model_by_model_size('eval/best_babyslm_test_syntactic', 'Syntactic Accuracy')
# plot_max_model_by_model_size('eval/best_babyslm_test_lexical', 'Lexical Accuracy', True)
# plot_max_model_by_model_size('eval/best_babyslm_test_syntactic', 'Syntactic Accuracy', True)
```

```python
def plot_max_model(eval_criteria, criteria_name, line=False):

    # Plot lexical accuracy by data size, taking the max dropout rate and max data size for each model size
    plt_data = df.groupby(['num_train_tokens'])[eval_criteria].max().reset_index()

    if line:
        # Line plot with log scale
        g = sns.lineplot(data=plt_data, x='num_train_tokens', y=eval_criteria)
        g.set(xscale="log")
        plt.title("Lexical Accuracy by Data Size")
        plt.show()

    else:
        # Bar plot with linear scale
        g = sns.catplot(data=plt_data, x="num_train_tokens", y=eval_criteria, kind="bar", height=6, aspect=1.5)
        g.set_axis_labels("Data Size", criteria_name)
        g.set(ylim=(0.4, 0.9))
        plt.title(f"{criteria_name} by Data Size")
        plt.show()
```

```python
plot_max_model('eval/best_babyslm_test_lexical', 'Lexical Accuracy')
plot_max_model('eval/best_babyslm_test_syntactic', 'Syntactic Accuracy')
```
